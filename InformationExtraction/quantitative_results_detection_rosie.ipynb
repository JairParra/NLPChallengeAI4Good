```python
quant_file = file.loc[file['Type de résultat'] == "Effet quantifié"]
#aggregate data from all years
columns = ['RESULTATS_2018', 'RESULTAT_2016', 'RESULTAT_2015', 'RESULTAT_2014','RESULTAT_2013', 'RESULTAT_2012', 'RESULTAT_2011']
data = pd.concat(file[header] for header in columns)
quant_data = [entry for entry in quant_file['RESULTATS_2018']]

#remove all NaN values
data = [entry for entry in data if str(entry) != 'nan']
tags = ["DET","ADP","PUNCT","CONJ","CCONJ","NUM","SYM","SPACE"]

data_tokenized = []
for entry in data:
    data_tokenized.append([token.lower_ for token in nlp(entry) if '%' in token.text or token.text.isdigit() or token.pos_ not in tags and token.text.isalpha()
                     and token.text not in stopwords.words('french')])
'''
stop_re = '\\b'+'\\b|\\b'.join(nltk.corpus.stopwords.words('french'))+'\\b'
quant_data = [re.sub(r"{}".format(stop_re), '', entry) for entry in quant_data]
quant_data = [unidecode(entry) for entry in quant_data]
'''
```

```python
total = 0
trend=0
word_bank = ['moyenne', 'augmentation', 'diminution', 'plus', 'moins', 'impact', 'lamoyenne', 'moyennne', 'moyennede', 'moyenn', 'hausse', 'baisse', 'accroissement', 'uneaugmentation', 'encore', 'très', 'suffisament', 'beaucoup', 'assez', 'excessivement', 'relativement', ]


for entry in data_tokenized:
    for token in entry:
        if token.isdigit():
            total += 1
            for word in word_bank:
                if word in entry[entry.index(token)-2:entry.index(token)+2]:
                    trend += 1 
                    print(' '.join(e for e in entry[entry.index(token)-2:entry.index(token)+2]))
                    
print(total)
print(trend)
```

```python
regexp = re.compile(r'((([\wÀ-ÿ])*(.){0,1}\s+)){0,10}(\d+([.]{0,3}\d+)*)((([\wÀ-ÿ])*(.){0,1}\s+)){0,10}.')
word_bank = ['individuel','individuelle','personnes','femmes','réfugiés', 'participant','participants','moyenne', 'augmentation', 'diminution', 'plus', 'moins', 'impact', 'lamoyenne', 'moyennne', 'moyennede', 'moyenn', 'hausse', 'baisse', 'accroissement', 'uneaugmentation', 'encore', 'très', 'suffisament', 'beaucoup', 'assez', 'excessivement', 'relativement', ]

file['RESULTATS_2018_matched'] = ""

window_list=[]

for index, row in file.iterrows():
    if str(row['RESULTATS_2018']) == 'nan':
        continue
    for x in re.finditer(regexp,str(row['RESULTATS_2018'])):
        window_list.append(x.group())
        for word in word_bank:
            if word in x.group().lower() and x.group() not in row['RESULTATS_2018_matched']:
                    row['RESULTATS_2018_matched']+= "\n" + x.group()
                    
                
```
